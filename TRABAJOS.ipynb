```python
df = df.drop(columns=["Links", "Unnamed: 0", "Date", "Description"])
```

```python
state_mapping = {
    ", CA": "CA",
    ", NY": "NY",
    ", TX": "TX",
    ", WI": "WI",
    ", WA": "WA",
    ", MA": "MA",
    ", MN": "MN",
    ", TN": "TN",
    ", IL": "IL",
    ", CT": "CT",
    ", MD": "MD",
    ", CO": "CO",
    ", NJ": "NJ",
    ", NC": "NC",
    ", GA": "GA",
    ", DC": "DC",
    ", OH": "OH",
    ", OR": "OR",
    ", VA": "VA",
    ", PA": "PA",
    ", FL": "FL",
    ", MI": "MI",
    ", MO": "MO",
    ", NE": "NE",
    ", AL": "AL",
    ", AR": "AR",
    ", AZ": "AZ",
    ", DE": "DE",
    ", IA": "IA",
    ", IN": "IN",
    ", KS": "KS",
    ", MS": "MS",
    ", NV": "NV",
    ", NM": "NM",
    ", OK": "OK",
    ", RI": "RI",
    ", SC": "SC",
    ", UT": "UT",
    "North Carolina": "NC",
    "South Carolina": "SC",
    "New Jersey": "NJ",
    "Texas": "TX",
    "Illinois": "IL",
    "California": "CA",
    "West Virginia": "WV",
    "remote": "Remote",
    "Remote": "Remote"
}
```

```python
def get_state(location):
    for key, value in state_mapping.items():
        if key in location:
            return value
    return "other loc"

df["Location"] = df["Location"].apply(get_state)
```

```python
df.head()
```

```python
df["Title"] = df["Title"].apply(str.lower)
```

```python
df.head()
```

```python
df["Title"] = df["Title"].apply(lambda x: "data engineer" if "data e" in x
                                          else "data science" if "data s" in x
                                          else "data analyst" if "data a" in x
                                          else "other data jobs" if "data" in x
                                          else "database" if "database" in x
                                          else "business" if "business" in x
                                          else "machine learning" if "machine" in x
                                          else "software engineer" if "software" in x
                                          else "developer" if "developer" in x
                                          else "other categories")
```

```python
df["Title"] = df["Title"].apply(str.capitalize)
```

```python
df.head()
```

```python
df["Salary"] = df["Salary"].str.replace("Up to", "-")
```

```python
df["Salary"].fillna("0", inplace=True)
```

```python
df["Hour/Year"] = df["Salary"].apply(lambda x: "Year" if "year" in x
                                     else "Month" if "month" in x
                                     else "Week" if "week" in x
                                     else "Hour" if "hour" in x
                                     else "Unknown")
```

```python
df[["Min Salary", "Max Salary"]] = df["Salary"].str.split("-", expand=True)
```

```python
df["Min Salary"] = df["Min Salary"].str.replace(r'[A-Za-z$]+', '', regex=True).str.strip()
```

```python
df["Max Salary"] = df["Max Salary"].str.replace(r'[A-Za-z$]+', '', regex=True).str.strip()
```

```python
df.head()
```

```python
df["Time Period"] = df["Hour/Year"]
```

```python
df.drop(columns=["Salary", "Hour/Year"], inplace=True)
```

```python
df.head()
```
